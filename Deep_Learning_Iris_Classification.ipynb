```python
import seaborn as sns
```

```python
df =  sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
dictionary = {'setosa':1, 'versicolor':2, 'virginica':3}
df['species'] = df['species'].map(dictionary)
```

```python
df.head()
```

```python
x = df.drop('species', axis=1)
```

```python
import tensorflow as tf
```

```python
y = tf.keras.utils.to_categorical(df['species'])
```

```python
inputlayer = tf.keras.layers.Dense(units=4, input_shape=(4,))
hiddenlayer1 = tf.keras.layers.Dense(units=32)
hiddenlayer2 = tf.keras.layers.Dense(units=32)
hiddenlayer3 = tf.keras.layers.Dense(units=64)
hiddenlayer4 = tf.keras.layers.Dense(units=64)
hiddenlayer5 = tf.keras.layers.Dense(units=128)
hiddenlayer6 = tf.keras.layers.Dense(units=64)
outputlayer = tf.keras.layers.Dense(units=4, activation='softmax')
```

```python
model = tf.keras.Sequential()
model.add(inputlayer)
model.add(hiddenlayer1)
model.add(hiddenlayer2)
model.add(hiddenlayer3)
model.add(hiddenlayer4)
model.add(hiddenlayer5)
model.add(hiddenlayer6)
model.add(outputlayer)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=10)
```
